```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# !pip install matplotlib  "instalo libreria"

# scatter plot "diagrama de dispersion"
x = np.arange(1,20)
y = x**2
#y1 = x**2 + 5**

plt.scatter(x,y,color='red',marker='s',label='x^2')
#plt.scatter(x,y1,color='blue',marker='s',label='x^2')
plt.legend(title='Funciones')
plt.title('Tipos de Funciones', fontsize=20)
plt.xlabel('x')
plt.ylabel('y')
plt.grid() #"agregar cuadrantes en el fondo"
# plt.xlim(0, 5) "modificar el rango de x"

plt.text(5, 300, "este texto", color='brown')

#agregar las coordenadas de los marcadores en texto
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x}, {val_x**2})')
```

```python
#Histogramas
media=20
std=3
numeros = np.random.normal(media, std, 1000)
plt.hist(numeros, bins=20, cumulative=True, density=True);
plt.title('Distribucion normal')
plt.xlabel('numeros')
plt.ylabel('frecuencia')
```
